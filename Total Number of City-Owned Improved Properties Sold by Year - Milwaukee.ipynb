## 2014 - 2017 profit analysis

### 2014 improved property sales

```python
# Third file analysis
props_2014 = pd.read_csv('2014-improved-property-sales.csv')
props_2014.head()
```

```python
sales_total_2014 = props_2014['Sale Price'].sum()
sales_total_2014
```

```python
total_mgt_2014 = props_2014['Total Mgt. Expenses'].sum().round()
total_mgt_2014
```

```python
totals_2014 = sales_total_2014 - total_mgt_2014
totals_2014
```
